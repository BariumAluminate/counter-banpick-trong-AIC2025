```python
%matplotlib inline
import numpy as np
import pandas as pd
from datascience import make_array
from datascience import Table
import datascience as ds
import matplotlib.pyplot as plt
```

```python
order = make_array(
    'BBan 1', 'Rban 1', 'BBan 2', 'Rban 2',
    'B1',
    'R1', 'R2',
    'B2', 'B3',
    'R3',
    'BBan 3', 'Rban 3','BBan 4', 'Rban 4',
    'R4',
    'B4', 'B5',
    'R5'
)

counter_slot = make_array(
     'B1',
     'R1', 'R2',
     'B2', 'B3',
     'R3',
     'R4',
     'B4', 'B5',
     'R5'
)
```

```python
def find_in_table(tbl, name):
    lb = tbl.labels
    ans = Table()
    for i in lb:
        ans = ans.with_column(i, make_array())
    for lb in counter_slot:
        temp = tbl.where(lb, name)
        ans.append(temp)
    return ans
```

```python
def counter_statistic(tbl : Table, name) :
    ans = Table().with_columns(
        'Name', make_array(),
        'count', make_array()
    )
    for i in range(len(counter_slot)):
        a = counter_slot[i]
        temp_tbl = tbl.where(counter_slot[i], name)
        for j in range(i + 1, len(counter_slot)):
            b = counter_slot[j]
            if b[0] == a[0]:
                continue
            temp = temp_tbl.group(counter_slot[j])
            temp.relabel(counter_slot[j], 'Name')
            ans.append(temp)
        ans = ans.group('Name', sum)
        ans.relabel('count sum', 'count')
    ans = ans.sort('count', descending=True)
    return ans
c = counter_statistic(ac, 'Marja')
c
```

```python
def replace_value_in_table(tbl, old_value, new_value):
    # Tạo một hàm phụ để áp dụng cho từng ô dữ liệu
    # Nếu gặp giá trị cũ thì đổi thành giá trị mới, ngược lại thì giữ nguyên
    def thay_the(x):
        return new_value if x == old_value else x

    # Tạo một bảng mới từ bảng gốc
    new_tbl = tbl

    # Duyệt qua từng cột trong bảng để cập nhật dữ liệu mới
    for lb in tbl.labels:
        # Dùng hàm .apply để biến đổi toàn bộ dữ liệu của cột 'lb'
        updated_column = new_tbl.apply(thay_the, lb)

        # Ghi đè cột đã sửa vào bảng mới
        new_tbl = new_tbl.with_column(lb, updated_column)

    return new_tbl

def remove_trailing_spaces(tbl):
    # Tạo bảng mới để lưu kết quả
    new_tbl = tbl

    # Định nghĩa một hàm phụ để xử lý từng ô dữ liệu
    def clean_cell(x):
        # Nếu ô đó là chuỗi chữ (string) thì mới cắt dấu cách,
        # nếu là số (int, float) thì giữ nguyên để tránh bị lỗi
        if isinstance(x, str):
            return x.rstrip()
        return x

    # Duyệt qua từng cột trong bảng để làm sạch
    for lb in tbl.labels:
        # Áp dụng hàm clean_cell cho toàn bộ cột 'lb'
        cleaned_column = new_tbl.apply(clean_cell, lb)

        # Ghi đè cột sạch vào bảng
        new_tbl = new_tbl.with_column(lb, cleaned_column)

    return new_tbl
AIC2025 = remove_trailing_spaces(AIC2025)
AIC2025
```

```python
def get_ban_statistic(tbl) :
    ans = Table().with_columns(
        'Name', make_array(),
        'count', make_array()
    )
    for label in tbl.labels:
        if 'Ban' in label or 'ban' in label:
            m = tbl.group(label)
            m.relabel(label, 'Name')
            ans.append(m)
    ans = ans.group('Name', sum)
    ans.relabel('count sum', 'count')
    ans = ans.sort('count', descending=True)
    return ans
ban_statistics = get_ban_statistic(AIC2025)
ban_statistics
```

```python
def get_pick_statistic(tbl) :
    ans = Table().with_columns(
        'Name', make_array(),
        'count', make_array()
    )
    for label in tbl.labels:
        if ('B' in label and 'Ban' not in label) or ('R' in label and 'ban' not in label):
            m = tbl.group(label)
            m.relabel(label, 'Name')
            ans.append(m)
    ans = ans.group('Name', sum)
    ans.relabel('count sum', 'count')
    ans = ans.sort('count', descending=True)
    return ans
pick_statistic = get_pick_statistic(AIC2025)
pick_statistic
```
